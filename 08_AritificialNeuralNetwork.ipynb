- perceptron

  단층 퍼셉트론은 xor gate 구현 불가 (직선만 표현 가능하기 때문에)

  image.png

```python
import torch
import torch.nn as nn
device = 'cuda'

X = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]]).to(device)
Y = torch.FloatTensor([[0],[1],[1],[0]]).to(device)

linear = nn.Linear(2,1,bias = True)
sigmoid = nn.Sigmoid()
model = nn.Sequential(linear,sigmoid).to(device)
optimizer = torch.optim.SGD(model.parameters(),lr=1)
c= torch.nn.BCELoss().to(device) # 이진분류에서 사용하는 크로스 엔트로피 함수

for step in range(10001):
  h = model(X)
  cost = c(h,Y)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if step % 100 == 0:
    print(step,cost.item())
```

- 인공신경망의 학습은 오차를 최소화하는 가중치를 찾는 목적으로 순전파와 역전파를 반복하는 것

- 다층 퍼셉트론 구현

```python
model = nn.Sequential(
    nn.Linear(2,10,bias = True), # input layer = 2
    nn.Sigmoid(),
    nn.Linear(10,10,bias = True), 
    nn.Sigmoid(),
    nn.Linear(10,10,bias = True), 
    nn.Sigmoid(),
    nn.Linear(10,1,bias = True), 
    nn.Sigmoid()
).to(device)

for epoch in range(10001):
    hypothesis = model(X)
    cost = c(hypothesis, Y)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(epoch, cost.item())
```
